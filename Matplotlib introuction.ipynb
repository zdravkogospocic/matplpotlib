# Which one should we use? ( pyplot vs matplotlib OO method

```python
heart
```

```python
over_50 = heart[heart["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
#Pyplot method
over_50.plot(kind="scatter", x ='age', y="chol",c ="target", figsize=(10,6));
```

```python
## OO method

fix, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age",y='chol', 
             c= 'target', ax=ax);

ax.set_xlim([45,100])
```

```python
fid, ax = plt.subplots(figsize=(10,6))

# Plot data

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c = over_50["target"])

# Customize

ax.set(title="Heart isease and Colesterol Levels",
      xlabel ="Age",
      ylabel ="Cholesterol")
# Add legend

ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line

ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

# Add data to ax0
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols =1,
                              figsize = (10,10))
scatter = ax0.scatter(x=over_50["age"],
                     y = over_50["chol"],
                     c= over_50["target"])
# Customize ax0

ax0.set(title="Heart diease and Cholesterol Levels",
       xlabel = "Age",
       ylabel="Cholesterol")
# Add a legend to ax0

ax0.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline

ax0.axhline(y= over_50["chol"].mean(), linestyle="--");

# Add data to ax1

scatter_1 = ax1.scatter(x=over_50["age"],
                     y = over_50["thalach"],
                     c= over_50["target"])
# Customize ax0

ax1.set(title="Heart diease and Max Heart Rate",
       xlabel = "Age",
       ylabel="Max heart rate")
# Add a legend to ax0

ax1.legend(*scatter.legend_elements(), title = "Target")

# Add a meanline

ax1.axhline(y= over_50["thalach"].mean(), linestyle="--");

# Add title to te figure

fig.suptitle("Hearte disease Analyses", fontsize = 16, fontweight="bold")
```
